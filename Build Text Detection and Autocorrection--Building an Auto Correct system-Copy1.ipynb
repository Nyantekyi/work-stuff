# Implementing the editing operations

```python
def split(word):
    return [(word[:i],word[i:]) for i in range(len(word) +1)]
    
```

```python
print(split('zzzzvancomycin inj. 50mg'))
```

```python
def delete(word):
    return [l + r[1:] for l,r in split(word) if r]
```

```python
print (delete('zzzzvancomycin inj. 50mg'))
```

```python
def swap(word):
    return [l + r[1] +r[0]+r[2:] for l,r in split(word) if len(r)>1]
```

```python
swap('paracetamol')
```

```python
def replace(word):
    letters = string.ascii_lowercase
    return [l + c +r[1:] for l,r in split(word) if r for c in letters]
```

```python
replace('paracetamol')
```

```python
def insert(word):
    letters= string.ascii_lowercase
    return [l + c+ r for l,r in split(word) for c in letters]
```

```python
insert('trash')
```

# Edit Distances

```python
def level_1_edits(word):
    return set(delete(word)+replace(word)+swap(word)+insert(word))
```

```python
level_1_edits('vancomycin inj. 50mg')
```

```python
def level_2_edit(word):
    return set(e2 for e1 in level_1_edits(word) for e2 in level_1_edits(e1))
```

```python
level_2_edit('vancomycin inj 50mg')
```

```python
def correct_spelling(word, vocabulary, word_probabilities):
    if word in vocabulary:
        print(f"{word} already correctly spelt")
    
    sugestions= level_1_edits(word) or level_2_edit(word) or [word]
    best_guesses = [w for w in sugestions if w in vocabulary]
    return [(w, word_probabilities[w]) for w in best_guesses]
```

```python
correct_spelling('Paracetamol',vocabs,word_probas)
```
