```python
import re
import string
from collections import Counter
import numpy as np
```

```python
 class spellchecker(object):
        def __init__(self, filename):
            with open(filename, "r",encoding='utf-8') as file:
                lines =file.readlines()
                words=[]

                for line in lines:
                    words += re.findall(r'[^\n]+', line.lower(),re.M)
                    words += re.findall(r'\w+', line.lower())
            self.vocabs= set(words)
            self.words_counts= Counter(words)
            self.total_word_count=float(sum(self.words_counts.values()))
            self.word_probas= {word: self.words_counts[word]/self.total_word_count for word in self.vocabs}
        def _level_1_edits(self,word):
            letters=string.ascii_lowercase
            splits=[(word[:i],word[i:]) for i in range(len(word) +1)]
            delete=[l + r[1:] for l,r in splits if r]
            swap=[l + r[1] +r[0]+r[2:] for l,r in splits if len(r)>1]
            replace=[l + c +r[1:] for l,r in splits if r for c in letters]
            insert=[l + c+ r for l,r in splits for c in letters]
            return set(delete+replace+swap+insert)
        def _level_2_edits(self,word):
            return set(e2 for e1 in self._level_1_edits(word) for e2 in self._level_1_edits(e1))
        def _level_3_edits(self,word):
            return set(e3 for e2 in self._level_2_edits(word) for e3 in self._level_2_edits(e2))
        def _level_4_edits(self,word):
            return set(e4 for e3 in self._level_3_edits(word) for e4 in self._level_3_edits(e3))
        def _level_5_edits(self,word):
            return set(e5 for e4 in self._level_4_edits(word) for e5 in self._level_4_edits(e4))
        def _level_6_edits(self,word):
            return set(e6 for e5 in self._level_5_edits(word) for e6 in self._level_5_edits(e5))
        
        def check(self,word):
            candidate= self._level_1_edits(word) or self._level_2_edits(word) or self._level_3_edits(word)or self._level_4_edits(word) or self._level_5_edits(word) or self._level_6_edits(word) or[word] 
            valid_candidate= [w for w in candidate if w in self.vocabs]
            #print(self.words)
            return sorted([(c,self.word_probas[c]) for c in valid_candidate],key=lambda tup:tup[1],reverse=True)
        #def check
checker=spellchecker('./New Text Document (2).txt')
```

# Implementing the editing operations

```python
def split(word):
    return [(word[:i],word[i:]) for i in range(len(word) +1)]
    
```

```python
print(split('paraceta'))
```

```python
def delete(word):
    return [l + r[1:] for l,r in split(word) if r]
```

```python
print (delete('paracetamol'))
```

```python
def swap(word):
    return [l + r[1] +r[0]+r[2:] for l,r in split(word) if len(r)>1]
```

```python
swap('paracetamol')
```

```python
def replace(word):
    letters = string.ascii_lowercase
    return [l + c +r[1:] for l,r in split(word) if r for c in letters]
```

```python
replace('paracetamol')
```

```python
def insert(word):
    letters= string.ascii_lowercase
    return [l + c+ r for l,r in split(word) for c in letters]
```

```python
insert('trash')
```

# Edit Distances

```python
def level_1_edits(word):
    return set(delete(word)+replace(word)+swap(word)+insert(word))
```

```python
level_1_edits('paracetamol')
```

```python
def level_2_edit(word):
    return set(e2 for e1 in level_1_edits(word) for e2 in level_1_edits(e1))
```

```python
level_2_edit('metformin')
```

```python
def correct_spelling(word, vocabulary, word_probabilities):
    if word in voc
    abulary:
        print(f"{word} already correctly spelt")
    
    sugestions= level_1_edits(word) or level_2_edit(word) or [word]
    best_guesses = [w for w in sugestions if w in vocabulary]
    return [(w, word_probabilities[w]) for w in best_guesses]
```
