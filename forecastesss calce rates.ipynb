```python

def rate_calculated_percentageincrease(This):
    if This['Min_col']=='AutoRegressive_False_mse':
        sale_target=((This['avg_sale_target']-This['Mean_Values'])/(This['Mean_Values']))
    elif This['Min_col']=='AutoRegressive_True_mse':
        sale_target=((This['avg_sale_target']-This['Mean_Values'])/(This['Mean_Values']))
    elif This['Min_col']=='Auto_Arima_Future_mse':
        sale_target=((This['avg_sale_target']-This['Mean_Values'])/(This['Mean_Values']))
    elif This['Min_col']=='Exponential_sm_mse':
        sale_target=((This['avg_sale_target']-This['Mean_Values'])/ (This['Mean_Values']))
    
    else:
        sale_target=np.NaN
    return sale_target

data['percentageIncrease']=data.apply(rate_calculated_percentageincrease,axis=1)
```
