```python
import pandas as pd
import textdistance as td
from difflib import SequenceMatcher
import nltk
from nltk.translate import bleu
from nltk.translate.bleu_score import SmoothingFunction
smoothie = SmoothingFunction().method4
```

```python
warehouse=pd.read_csv('/Users/macos/Downloads/ProductStocklist-lava.csv')[['id**', 'name*']].drop_duplicates()
```

```python

ite_=[]
item_=[]

hamming_=[]
hamming_norm_=[]
leven_=[]
leven_norm_=[]
damerau_=[]
damerau_norm_=[]
jaro_=[]
jarowin_=[]
smith_=[]

jaccard_=[]

tvers_=[]
over_=[]
cosine_=[]
seqmat_=[]
smoo_=[]


for item in warehouse['name*'].unique():
    print(item)
    print('_________________________________')
    tversky = td.Tversky(ks=(1, 1))
    
    for ite in warehouse['name*'].unique():
        hamming=td.hamming(str(item), str(ite))
        hamming_norm=td.hamming.normalized_similarity(str(item), str(ite))

        leven=td.levenshtein(str(item), str(ite))
        leven_norm=td.levenshtein.normalized_similarity(str(item), str(ite))
        
        damerau=td.damerau_levenshtein(str(item), str(ite))
        damerau_norm=td.damerau_levenshtein.normalized_similarity(str(item), str(ite))
        jaro=td.jaro(str(item), str(ite))
        jarowin=td.jaro_winkler(str(item), str(ite))
        smith=td.smith_waterman(str(item), str(ite))
        jaccard=td.jaccard(str(item).split(), str(ite).split())
#         sorencen=td.sorencen(str(item).split(), str(ite).split())

        tvers=tversky(str(item).split(), str(item).split())
        over=td.overlap(str(item).split(), str(item).split())
        cosine=td.cosine(str(item).split(), str(item).split())
        seqmat=SequenceMatcher(None, item, ite).ratio()
        smoo=bleu([str(item)], str(ite), smoothing_function=smoothie)        
        
        
        
        ite_.append(ite)


        item_.append(item)

        hamming_.append(hamming)
        hamming_norm_.append(hamming_norm)
        leven_.append(leven)
        leven_norm_.append(leven_norm)
        damerau_.append(damerau)
        damerau_norm_.append(damerau_norm)
        jaro_.append(jaro)
        jarowin_.append(jarowin)
        smith_.append(smith)

        jaccard_.append(jaccard)

        tvers_.append(tvers)
        over_.append(over)
        cosine_.append(cosine)
        seqmat_.append(seqmat)
        smoo_.append(smoo)

        
        
        

details = {
    'item1':ite_,'item2':item_,'hamming':hamming_,'hamming_norm':hamming_norm_,'leven':leven_,
    'leven_norm':leven_norm_,'damerau':damerau_,'damerau_norm':damerau_norm_,'jaro':jaro_,
    'jarowin':jarowin_,'smith':smith_,'jaccard':jaccard_,'tvers':tvers_,'over':over_,'cosine':cosine_,
    'seqmat':seqmat_,'smoo':smoo_
}
 
# creating a Dataframe object 
df = pd.DataFrame(details)
        
```
